```python
# =============================================================================
# NOTEBOOK 6: EXPERIMENTO 3 - ROBUSTEZ Y AN√ÅLISIS PROFUNDO (FULL)
# =============================================================================
# 1. Performance con datos fuera de distribuci√≥n (Simulado con Ruido)
# 2. An√°lisis de errores por categor√≠a (Confusion Matrix)
# 3. Sensibilidad a hiperpar√°metros (Dropout) PARA TODAS LAS ARQUITECTURAS
# 4. CONCLUSI√ìN FINAL Y RANKING
# =============================================================================

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import pandas as pd
import numpy as np
import pickle
import os
import random
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score

# --- CONFIGURACI√ìN ---
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
DATA_DIR = "data_processed"
MODELS_DIR = "models"
SEED = 42

random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)

print(f"‚öôÔ∏è Usando dispositivo: {DEVICE}")

# -----------------------------------------------------------------------------
# DEFINICI√ìN DE CLASES
# -----------------------------------------------------------------------------
class AttentionPooling(nn.Module):
    def __init__(self, hidden_dim):
        super().__init__()
        self.attn = nn.Linear(hidden_dim, 1)
    def forward(self, rnn_output, mask):
        scores = self.attn(rnn_output).squeeze(-1)
        scores = scores.masked_fill(mask == 0, -1e9)
        attn_weights = torch.softmax(scores, dim=1)
        context = torch.sum(rnn_output * attn_weights.unsqueeze(-1), dim=1)
        return context, attn_weights

class RecurrentClassifier(nn.Module):
    def __init__(self, model_type, vocab_size, embed_dim, hidden_dim, out_dim, n_layers, dropout, pad_idx, bidirectional=False, use_attention=True):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=pad_idx)
        self.model_type = model_type
        self.bidirectional = bidirectional
        self.use_attention = use_attention
        self.n_layers = n_layers
        self.hidden_dim = hidden_dim
        
        if model_type == 'lstm':
            self.rnn = nn.LSTM(embed_dim, hidden_dim, n_layers, batch_first=True, dropout=dropout if n_layers > 1 else 0, bidirectional=bidirectional)
        else:
            self.rnn = nn.GRU(embed_dim, hidden_dim, n_layers, batch_first=True, dropout=dropout if n_layers > 1 else 0, bidirectional=bidirectional)
        
        self.rnn_out_dim = hidden_dim * (2 if bidirectional else 1)
        if use_attention: self.attention = AttentionPooling(self.rnn_out_dim)
        else: self.attention = None
        self.fc = nn.Linear(self.rnn_out_dim, out_dim)
        self.dropout = nn.Dropout(dropout)
        self.pad_idx = pad_idx

    def forward(self, x, lengths=None):
        mask = (x != self.pad_idx).float()
        emb = self.embedding(x)
        if self.model_type == 'lstm': rnn_out, _ = self.rnn(emb)
        else: rnn_out, _ = self.rnn(emb)
        
        attn_weights = None
        if self.use_attention: context, attn_weights = self.attention(rnn_out, mask)
        else:
            if self.bidirectional: context = torch.mean(rnn_out, dim=1)
            else: context = rnn_out[:, -1, :]
        return self.fc(self.dropout(context)), attn_weights

# -----------------------------------------------------------------------------
# CARGA DE DATOS
# -----------------------------------------------------------------------------
print("üì• Cargando datos y vocabulario...")
vocab = pickle.load(open(f"{DATA_DIR}/vocab.pkl", "rb"))
test_df = pd.read_csv(f"{DATA_DIR}/test.csv")
PAD_IDX, UNK_IDX = vocab["<PAD>"], vocab["<UNK>"]
labels_map = {0: 'Bajista', 1: 'Alcista', 2: 'Neutral'}

def encode_text(text, vocab):
    tokens = str(text).lower().split()
    ids = [vocab.get(tok, UNK_IDX) for tok in tokens]
    return torch.tensor(ids, dtype=torch.long)

def add_noise(text, prob=0.1):
    chars = list(text)
    for i in range(len(chars)):
        if random.random() < prob:
            if random.random() < 0.5: chars[i] = '' 
            else: chars[i] = chr(random.randint(97, 122))
    return "".join(chars)

# =============================================================================
# PARTES 1 Y 2: AN√ÅLISIS DE MODELOS GUARDADOS (Robustez y Errores)
# =============================================================================
model_files = [
    'lstm_base_best_model.pth', 'lstm_attn_best_model.pth',
    'gru_base_best_model.pth', 'gru_attn_best_model.pth',
    'lstm_bi_best_model.pth', 'lstm_bi_attn_best_model.pth'
]

summary_metrics = [] 

print(f"\nüöÄ [FASES 1 y 2] AN√ÅLISIS DE MODELOS PRE-ENTRENADOS ({len(model_files)})...\n")

for m_file in model_files:
    path = os.path.join(MODELS_DIR, m_file)
    if not os.path.exists(path): continue

    print("="*60)
    print(f" ü§ñ MODELO: {m_file.replace('_best_model.pth', '').upper()}")
    print("="*60)

    # Cargar
    checkpoint = torch.load(path, map_location=DEVICE)
    cfg = checkpoint['config']
    model = RecurrentClassifier(
        cfg['model_type'], cfg['vocab_size'], cfg['embed_dim'], cfg['hidden_dim'], 3,
        cfg['n_layers'], cfg['dropout'], cfg['pad_idx'], cfg.get('bidirectional', False), cfg.get('use_attention', True)
    ).to(DEVICE)
    model.load_state_dict(checkpoint['model_state'])
    model.eval()

    # Predicciones
    all_preds, all_labels = [], []
    for i in range(len(test_df)):
        text = test_df.iloc[i]['text']
        label = test_df.iloc[i]['label']
        tensor = encode_text(text, vocab).unsqueeze(0).to(DEVICE)
        with torch.no_grad():
            logits, _ = model(tensor)
            pred = logits.argmax(1).item()
        all_preds.append(pred)
        all_labels.append(label)

    clean_acc = accuracy_score(all_labels, all_preds)
    clean_f1 = f1_score(all_labels, all_preds, average='macro')
    
    # Reporte
    print(f"\nüìä Reporte (F1: {clean_f1:.4f}):")
    print(classification_report(all_labels, all_preds, target_names=labels_map.values()))

    # Matriz
    cm = confusion_matrix(all_labels, all_preds)
    plt.figure(figsize=(4, 3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels_map.values(), yticklabels=labels_map.values())
    plt.title(f"Confusion Matrix: {m_file[:-15]}")
    plt.show()

    # Robustez
    print(f"\nüå™Ô∏è Test de Robustez:")
    noise_levels = [0.0, 0.05, 0.1, 0.2]
    robustness_acc = []
    acc_noisy_10 = 0 

    for nl in noise_levels:
        noisy_preds = []
        for text in test_df['text']:
            noisy_text = add_noise(text, prob=nl)
            tensor = encode_text(noisy_text, vocab).unsqueeze(0).to(DEVICE)
            with torch.no_grad():
                pred = model(tensor)[0].argmax(1).item()
            noisy_preds.append(pred)
        acc = accuracy_score(test_df['label'], noisy_preds)
        robustness_acc.append(acc)
        if nl == 0.1: acc_noisy_10 = acc
        print(f"   - Ruido {nl*100}% -> Acc: {acc:.4f}")

    plt.figure(figsize=(4, 2))
    plt.plot([str(int(n*100))+'%' for n in noise_levels], robustness_acc, marker='o', linestyle='--', color='red')
    plt.title(f"Robustez: {m_file[:-15]}")
    plt.grid(True)
    plt.show()
    
    summary_metrics.append({
        'Modelo': m_file.replace('_best_model.pth', ''),
        'Macro_F1': clean_f1,
        'Acc_Clean': clean_acc,
        'Acc_Ruido_10%': acc_noisy_10
    })
    print("\n" + "-"*60 + "\n")

# =============================================================================
# PARTE 3: SENSIBILIDAD A HIPERPAR√ÅMETROS (TODAS LAS ARQUITECTURAS)
# =============================================================================
print("="*60)
print("üéõÔ∏è [FASE 3] AN√ÅLISIS DE SENSIBILIDAD COMPARATIVO (DROPOUT)")
print("="*60)
print("Entrenando variantes de las 6 arquitecturas (esto tomar√° unos minutos)...")

# Configuraciones de las 6 arquitecturas
configs = [
    {'name': 'LSTM Base', 'type': 'lstm', 'bi': False, 'attn': False},
    {'name': 'LSTM Attn', 'type': 'lstm', 'bi': False, 'attn': True},
    {'name': 'GRU Base',  'type': 'gru',  'bi': False, 'attn': False},
    {'name': 'GRU Attn',  'type': 'gru',  'bi': False, 'attn': True},
    {'name': 'BiLSTM',    'type': 'lstm', 'bi': True,  'attn': False},
    {'name': 'BiLSTM Attn','type':'lstm', 'bi': True,  'attn': True}
]

dropout_values = [0.1, 0.3, 0.5, 0.7]
train_df = pd.read_csv(f"{DATA_DIR}/train.csv") 

# Dataset temporal
class TempDS(Dataset):
    def __init__(self, df, vocab): self.df, self.vocab = df, vocab
    def __len__(self): return len(self.df)
    def __getitem__(self, i): 
        return encode_text(self.df.iloc[i]['text'], self.vocab).long(), self.df.iloc[i]['label']

def temp_collate(batch):
    x, y = zip(*batch)
    mx = max([len(s) for s in x])
    padded = [torch.cat([s.long(), torch.tensor([PAD_IDX]*(mx-len(s)), dtype=torch.long)]) for s in x]
    return torch.stack(padded), torch.tensor(y, dtype=torch.long)

# Dataloaders reducidos (30% datos, 3 epochs)
t_loader = DataLoader(TempDS(train_df.sample(frac=0.3, random_state=SEED), vocab), batch_size=32, collate_fn=temp_collate)
v_loader = DataLoader(TempDS(pd.read_csv(f"{DATA_DIR}/val.csv"), vocab), batch_size=32, collate_fn=temp_collate)

# Diccionario para guardar resultados de todas las arquitecturas
all_sensitivity_results = {}

for conf in configs:
    model_name = conf['name']
    print(f"\n--> Probando sensibilidad para: {model_name}")
    results_for_model = []
    
    for d in dropout_values:
        # Instanciar modelo nuevo
        model = RecurrentClassifier(
            model_type=conf['type'],
            vocab_size=len(vocab),
            embed_dim=100,
            hidden_dim=64,
            out_dim=3,
            n_layers=1, # Capas reducidas para velocidad
            dropout=d,
            pad_idx=PAD_IDX,
            bidirectional=conf['bi'],
            use_attention=conf['attn']
        ).to(DEVICE)
        
        opt = torch.optim.Adam(model.parameters(), lr=0.001)
        crit = nn.CrossEntropyLoss()
        
        # Mini-Entrenamiento (3 Epochs)
        model.train()
        for _ in range(3):
            for bx, by in t_loader:
                bx, by = bx.to(DEVICE), by.to(DEVICE)
                opt.zero_grad()
                logits, _ = model(bx)
                loss = crit(logits, by)
                loss.backward()
                opt.step()
        
        # Validaci√≥n
        model.eval()
        correct, total = 0, 0
        with torch.no_grad():
            for bx, by in v_loader:
                bx, by = bx.to(DEVICE), by.to(DEVICE)
                preds = model(bx)[0].argmax(1)
                correct += (preds == by).sum().item()
                total += by.size(0)
        
        acc = correct/total if total > 0 else 0
        results_for_model.append(acc)
        print(f"    Dropout {d} -> Val Acc: {acc:.4f}")
        
    all_sensitivity_results[model_name] = results_for_model

# GR√ÅFICO COMPARATIVO DE SENSIBILIDAD
plt.figure(figsize=(10, 6))
markers = ['o', 's', '^', 'D', 'v', 'X']
for i, (name, res) in enumerate(all_sensitivity_results.items()):
    plt.plot(dropout_values, res, marker=markers[i], label=name, linewidth=2)

plt.title("Comparaci√≥n de Sensibilidad al Dropout (Todas las Arquitecturas)")
plt.xlabel("Dropout Rate")
plt.ylabel("Validation Accuracy")
plt.legend()
plt.grid(True)
plt.show()

# =============================================================================
# PARTE 4: CONCLUSI√ìN FINAL Y RANKING
# =============================================================================
print("\n" + "="*60)
print("üèÜ RANKING FINAL Y VEREDICTO")
print("="*60)

if summary_metrics:
    df_res = pd.DataFrame(summary_metrics)
    df_res = df_res.sort_values(by='Macro_F1', ascending=False).reset_index(drop=True)

    print("\nüìã TABLA DE POSICIONES (Ordenado por Macro F1):")
    print(df_res.to_string(index=False))

    winner = df_res.iloc[0]
    most_robust = df_res.sort_values(by='Acc_Ruido_10%', ascending=False).iloc[0]

    print("\nüó£Ô∏è EXPLICACI√ìN DEL RESULTADO:")
    print(f"1. EL MEJOR MODELO (CAMPE√ìN): {winner['Modelo'].upper()}")
    print(f"   - Gan√≥ porque tiene el mejor equilibrio entre clases (Macro F1: {winner['Macro_F1']:.4f}).")

    if winner['Modelo'] == most_robust['Modelo']:
        print(f"   - ¬°Doble victoria! Tambi√©n es el m√°s resistente al ruido (Acc con ruido: {winner['Acc_Ruido_10%']:.4f}).")
    else:
        print(f"\n2. MENCI√ìN HONOR√çFICA A LA ROBUSTEZ: {most_robust['Modelo'].upper()}")
        print(f"   - Aunque {winner['Modelo']} es m√°s preciso en general, el modelo {most_robust['Modelo']}")
        print(f"     soport√≥ mejor la mala ortograf√≠a.")

    print("\nüí° CONCLUSI√ìN SOBRE SENSIBILIDAD:")
    print("   El gr√°fico comparativo muestra qu√© arquitecturas son m√°s estables.")
    print("   Generalmente, las arquitecturas con Atenci√≥n (Attn) toleran mejor el dropout alto")
    print("   porque pueden enfocarse en palabras clave aunque se apaguen neuronas aleatorias.")
```
